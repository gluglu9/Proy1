```python

import numpy as np
import pandas as pd

# pregunta 1


# Inicializar listas para almacenar los datos
dias = []
productos = []
unidades = []
precios = []

# Pedir al usuario los datos de ventas de los últimos 5 días
for i in range(5):
    dia = input(f"Introduce el día {i+1}: ")
    producto = input(f"Introduce el producto vendido el día {i+1}: ")
    unidad = int(input(f"Introduce el número de unidades vendidas el día {i+1}: "))
    precio = float(input(f"Introduce el precio unitario del producto el día {i+1}: "))

    dias.append(dia)
    productos.append(producto)
    unidades.append(unidad)
    precios.append(precio)

# Crear el DataFrame
data = {
    'Día': dias,
    'Producto': productos,
    'Unidades': unidades,
    'Precio Unitario': precios
}
df = pd.DataFrame(data)
df.set_index('Día', inplace=True)

# Calcular el precio con descuento
df['Precio con Descuento'] = df['Precio Unitario'] * 0.95

# Mostrar el DataFrame
print("\nDatos de ventas:")
print(df)



# -----------------------------------------------------------------------------
# -----------------------------------------------------------------------------
# PREGUNTA 2


# Generar ingresos aleatorios para 10 clientes
np.random.seed(0)  # Para reproducibilidad
clientes = [f'Cliente {i+1}' for i in range(10)]
ingresos = {cliente: np.random.randint(2000, 10000) for cliente in clientes}

# Crear una Serie de Pandas a partir del diccionario
serie_ingresos = pd.Series(ingresos)

# Calcular los estadísticos
estadisticos = pd.Series({
    'Ingreso Mínimo': serie_ingresos.min(),
    'Ingreso Máximo': serie_ingresos.max(),
    'Ingreso Medio': serie_ingresos.mean()
})

# Mostrar la serie con los estadísticos
print(ingresos)
print(estadisticos)




# -----------------------------------------------------------------------------
# -----------------------------------------------------------------------------
# PREGUNTA 3


# Crear el DataFrame con los datos proporcionados
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Ventas': [30500, 35600, 28300, 33900, 42500],
    'Gastos': [22000, 23450, 18100, 35700, 32450]
}
df = pd.DataFrame(data)
df.set_index('Mes', inplace=True)

# Generar nuevas columnas
df['Gasto/Ventas (%)'] = (df['Gastos'] / df['Ventas']) * 100
df['Meta'] = df['Ventas'].apply(lambda x: 'Meta Superada' if x >= 30000 else 'Meta no Superada')
df['Sobre Gasto'] = df.apply(lambda row: row['Gastos'] - row['Ventas'] if row['Gastos'] > row['Ventas'] else 0, axis=1)

# Obtener el total de las ventas de los últimos 3 meses
total_ventas_ultimos_3_meses = df['Ventas'][-3:].sum()

# Obtener el total de gastos de febrero a abril
total_gastos_febrero_abril = df.loc['Febrero':'Abril', 'Gastos'].sum()

# Mostrar el DataFrame y los totales calculados
print(df)
print(f"\nTotal de ventas de los últimos 3 meses: {total_ventas_ultimos_3_meses}")
print(f"Total de gastos de febrero a abril: {total_gastos_febrero_abril}")



# -----------------------------------------------------------------------------
# -----------------------------------------------------------------------------
# PREGUNTA 4


from google.colab import files

# Subir el archivo CSV
uploaded = files.upload()

# Asumiendo que el archivo subido se llama 'tu_archivo.csv'
nombre_archivo = 'datos_seguros.csv'

# 1. Generar un DataFrame con los datos del fichero
df = pd.read_csv(nombre_archivo)

# 2. Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los nombres de sus columnas y filas, los tipos de datos de las columnas, mostrar las 10 primeras filas y las 10 últimas filas, verificar si tiene datos perdidos/nulos
print("Dimensiones del DataFrame:", df.shape)
print("Número de datos:", df.size)
print("Nombres de las columnas:", df.columns.tolist())
print("Tipos de datos de las columnas:\n", df.dtypes)
print("Primeras 10 filas:\n", df.head(10))
print("Últimas 10 filas:\n", df.tail(10))
print("Datos perdidos/nulos:\n", df.isnull().sum())

# 3. Extraer las 1000 primeras filas y almacenarlas en un contenedor
primeras_1000_filas = df.head(1000)

# Mostrar las primeras 5 filas del contenedor para verificar
print("Primeras 5 filas del contenedor:\n", primeras_1000_filas.head())


1

```
